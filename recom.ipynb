```python
# JSON to DataFrame
df = pd.json_normalize(result['items'])
```

```python
# html 태그 제거, 이상한 문자 제거

def cleanhtml(raw) :
    p = re.compile('<.*?>')
    cleantxt = re.sub(p, '', raw)
    cleantxt = cleantxt.replace('&#x0D','')
    return cleantxt
```

```python
df['title'] = df['title'].apply(cleanhtml)
df['description'] = df['description'].apply(cleanhtml)
```

```python
df = df[['title','description']]
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
# TF - IDF Matrix 생성

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['description'].values.astype('U'))
print(tfidf_matrix.shape)
print(tfidf_matrix)
```

```python
# 결측치 확인 - 없음
df.isnull().sum()
```

```python
# 코사인 유사도 계산하기
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

```python
title = '두근두근 파이썬 (쉽고! 재밌고! 흥미로운 코딩 맛!)' # 28번째 도서와 다른 책의 유사도 체크
title2index = pd.Series(data = df.index, index = df['title']).drop_duplicates()
index = title2index[title]
```

```python
index
```

```python
# index 번째 도서에 대한 유사도 값에 대해 [(0, 유사도), (1, 유사도), ...] 형태로 저장
sim_scores = list(enumerate(cosine_sim[index]))

sim_scores
```

```python
# 유사도 기준으로 정렬하기
sim_scores.sort(key = lambda x:x[1], reverse = True)
top10 = sim_scores[1:11]
top10_index = [x[0] for x in top10]
print(top10_index)
```

```python
# 해당 책과 가장 유사한 책 10권을 추천
print(df['title'].iloc[top10_index])
```
